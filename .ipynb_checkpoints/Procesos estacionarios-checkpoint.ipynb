# Procesos estacionarios

```python
import os #Útil para acceder a los directorios
import pandas as pd #Manejo de base de datos
from functools import reduce #Función para vectorizar una función
from statsmodels.tsa.stattools import adfuller, kpss #Pruebas de estacionariedad
#Gráficos
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid') #Estilo
from statsmodels.tsa.seasonal import seasonal_decompose #Estacionalidad
import numpy as np #Matemática
```

```python
# Pruebas de estacionariedad
def adf_test(timeseries, regression = "c"):
    print("Prueba de Dickey-Fuller Aumentada:")
    dftest = adfuller(timeseries, autolag="AIC", regression = regression)
    dfoutput = pd.Series(
        dftest[0:4],
        index=[
            "Estadístico",
            "p-valor",
            "#Rezagos usados",
            "Número de observaciones",
        ],
    )
    for key, value in dftest[4].items():
        dfoutput["Valor crítico (%s)" % key] = value
    print(dfoutput)


def kpss_test(timeseries, regression = "c"):
    print("KPSS:")
    kpsstest = kpss(timeseries, regression= regression, nlags="auto")
    kpss_output = pd.Series(
        kpsstest[0:3], index=[
            "Estadístico",
            "p-valor",
            "#Rezagos usados"]
    )
    for key, value in kpsstest[3].items():
        kpss_output["Valor crítico (%s)" % key] = value
    print(kpss_output)
```

## Caminata aleatoria

```python
np.random.seed(54)
epsilon = np.random.standard_normal(1000)
epsilon[0] = 0
caminata = pd.Series(np.cumsum(epsilon))
plt.plot(caminata);
```

```python
adf_test(caminata)
```

```python
kpss_test(caminata)
```

```python
errores = caminata.diff()
#adf_test(errores)
```

```python
errores = caminata.diff().dropna()
adf_test(errores)
```

## Caso real

```python
carpeta_base = r'C:\Users\josea\Documents\GitHub\curso_series_tiempo\Datos'
# Leemos la base
serie_ipc = pd.read_excel(os.path.join(carpeta_base, "IPC.xlsx"), skiprows=4)
# Convertimos los meses en español por números
meses = (("Enero", "1"), ("Febrero", "2"), ("Marzo", "3"), ("Abril", "4"), 
         ("Mayo", "5"), ("Junio", "6"), ("Julio","7"), ("Agosto", "8"), 
         ("Septiembre", "9"), ("Octubre", "10"), ("Noviembre", "11"), 
         ("Diciembre", "12"))
serie_ipc.columns = ["Mes", "Nivel", "Var_Mensual", "Var_IA", "Var_Acum"]
serie_ipc['Mes'] = serie_ipc['Mes'].apply(lambda x: reduce(
    lambda a, kv: a.replace(*kv), meses, x))
#Damos formato de fecha
serie_ipc['Mes'] = pd.to_datetime(serie_ipc['Mes'],format = "%m/%Y")
serie_ipc.index = serie_ipc['Mes'] 
serie_ipc.drop(columns = ['Mes'], inplace = True)
```

```python
#Principales estadísticas
serie_ipc.describe()
```

### Visualización

```python
sns.lineplot(x="Mes", y = "Nivel",data = serie_ipc);
plt.title("Índice de precios al consumidor");
plt.xlabel('');
```

```python
sns.lineplot(x="Mes", y = "Var_Mensual",data = serie_ipc);
plt.title("IPC, variación mensual");
plt.ylabel("Porcentaje (%)");
plt.xlabel('');
```

```python
sns.lineplot(x="Mes", y = "Var_IA",data = serie_ipc);
plt.title("IPC, variación interanual");
plt.ylabel("Porcentaje (%)");
plt.xlabel('');
```

### Pruebas estadísticas

```python
#Sobre la serie original
adf_test(serie_ipc['Nivel'], regression = "c")
```

```python
#Sobre la variación mensual
adf_test(serie_ipc['Var_Mensual'].dropna(), regression = "c")
```

```python
#Sobre la variación interanual
adf_test(serie_ipc['Var_IA'].dropna(), regression = "c")
```
